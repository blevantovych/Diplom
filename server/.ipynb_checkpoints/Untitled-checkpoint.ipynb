```python
def expandInterval(interval, history):
    [start, end] = interval
    if (start > end):
        print('Begining of interval is greater than its end')
        return
    if len(history) == 0:
        print('when expanding there should be history')
        return

    nearest_right_neighbor = history[0][1]
    for i in history:
        if i[1] > end and i[1] < nearest_right_neighbor:
            nearest_right_neighbor = i[1]
    delta = (nearest_right_neighbor - end) / 2
    return [start, end + delta]
```

```python
expandInterval([0, 4.375], [[0, 5], [0, 3.75]])
```

```python
def shrinkInterval(interval, history=[]):
    [start, end] = interval
    if (start > end):
        print('Begining of interval is greater than its end')
        return
    if len(history) == 0:
        mid = (end - start) / 2.0
        return [start, mid]

    nearest_left_neighbor = 0
    for i in history:
        if i[1] < end and i[1] > nearest_left_neighbor:
            nearest_left_neighbor = i[1]
    delta = (end - nearest_left_neighbor) / 2
    return [start, end - delta]
```

```python
shrinkInterval([0, 2.34375], [[0, 5], [0, 2.5], [0, 1.25], [0, 1.875], [0, 2.1875], [0, 2.34375]])
```
